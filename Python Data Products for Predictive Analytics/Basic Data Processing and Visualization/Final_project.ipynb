I have used the amazon gift card data set. This data set seems to be well organized and easy to work with.

```python
import numpy
import matplotlib.pyplot as plt
import csv
import gzip
from collections import defaultdict
```

```python
path = 'amazon_reviews_us_Gift_Card_v1_00.tsv.gz'
f = gzip.open(path, 'rt')
reader = csv.reader(f, delimiter='\t')
```

```python
header = next(reader)
header
```

```python
# Attatch the header to the lines of data
dataset = []
for line in reader:
    d = dict(zip(header, line))
    dataset.append(d)
```

```python
dataset[0]
```

```python
# Length of dataset

length = len(dataset)
length
```

```python
# average rating

all_ratings = [int(d['star_rating']) for d in dataset]
sum_ratings = sum(all_ratings)
average_rating = sum_ratings / length
average_rating
```

```python
# find range of ratings

all_ratings = [int(d['star_rating']) for d in dataset]
sorted_all_ratings = sorted(all_ratings)
lowset_rating = sorted_all_ratings[0]
highest_rating = sorted_all_ratings[-1]

range_of_ratings = highest_rating - lowset_rating
range_of_ratings
```

```python
product = [d['product_id'] for d in dataset]
product
```

```python
ratings = defaultdict(int)
```

```python
for d in dataset:
    if float(d['star_rating']) >0 and float(d['star_rating']) <=1:
        ratings[1] += 1
    elif float(d['star_rating']) >1 and float(d['star_rating']) <=2:
        ratings[2] += 1
    elif float(d['star_rating']) >2 and float(d['star_rating']) <=3:
        ratings[3] += 1
    elif float(d['star_rating']) >3 and float(d['star_rating']) <=4:
        ratings[4] += 1
    else:
        ratings[5] += 1
```

```python
rate = list(ratings.keys())
count = list(ratings.values())
plt.xlabel('Rating')
plt.ylabel('Count of Rating')
plt.title('Distribution of Ratings')
plt.bar(rate,count)
```
